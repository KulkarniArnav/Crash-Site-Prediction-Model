```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn import metrics 
from sklearn.metrics import classification_report, confusion_matrix

X = df1[['ACCIDENT_date','ACCIDENT_month','ACCIDENT_HOUR','ALCOHOL_RELATED','SPEED_ZONE']]
y = df1[['LATITUDE','LONGITUDE']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)

lm = LinearRegression()
lm.fit(X_train,y_train)

test_val = [5, 2, 9, 1, 100]

test_prediction = lm.predict(np.array(test_val).reshape(1, -1))
test_prediction = test_prediction.astype("float")

test_lat = test_prediction[0][0]
test_long = test_prediction[0][1]

print(test_lat,"\n")
print(test_long)


import folium

map = folium.Map(location = [test_lat,test_long], zoom_start=10)

folium.Marker([test_lat,test_long], popup = '<strong>Test Marker</strong>').add_to(map)

map
```
